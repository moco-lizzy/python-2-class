```python
from abc import ABC, abstractmethod


class Product(ABC):
    category = 'Product'
    @abstractmethod
    def __init__ (self, price, item_id, quantity):
        pass
    def get_info(self):
        pass
    def get_category(self):
        pass

class Book(Product):
    category = 'Book'
    def __init__ (self, price, item_id, quantity, author, title):
        self.price = price
        self.item_id = item_id
        self.quantity = quantity
        self.author = author
        self.title = title

    def get_info(self):
        return f"{self.category}\t {self.price}\t{self.item_id}\t{self.quantity}\t{self.author}\t{self.title}"
    def get_category(self):
        return f"Category: {self.category}"

class Movie(Product):
    category = 'Movie'
    def __init__ (self, price, item_id, quantity, title, genre, year):
        self.price = price
        self.item_id = item_id
        self.quantity = quantity
        self.title = title
        self.__genre = genre
        self.__year = year
    
    def get_info(self):
        return f"{self.category}\n{self.price}\n{self.item_id}\n{self.quantity}\n{self.__genre}\n{self.title}\n{self.__year}"     
    def get_category(self):
        return f"Category: {self.category}"
    
class Inventory():
    def __init__ (self):
        self.__list = []
    def add_item (self, item):
        self.__list.append(item)
    def show_inventory(self):
        msg = ""
        index = 1
        for item in self.__list:
            msg += str(index) + ". "    + item.get_info() + "\n"
            index += 1 
        return msg
        

        
book1 = Book(12,2190, 13,"Hemingway", "The Sun Also Rises")

movie1 = Movie(20, 2102, 30, "Alien","Horror",1980)

inventory_list = Inventory()

inventory_list.add_item(book1)
inventory_list.add_item(movie1)

print(inventory_list.show_inventory())



```
