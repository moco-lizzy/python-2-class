```python
from abc import ABC, abstractmethod





class Product(ABC):
    category = 'Product'
    @abstractmethod
    def __init__ (self, item_id, price, quantity):
        pass
    def get_info(self):
        pass
    def get_category(self):
        pass

class Book(Product):
    category = 'Book'
    def __init__ (self,item_id, title, price, quantity):
        self.price = price
        self.item_id = item_id
        self.quantity = quantity
        self.title = title
        self.__author = None
    def get_info(self):
        return f"|{self.category} |\t{self.item_id}|\t{self.title}|\t{self.price}|\t{self.quantity}|\t{self.__author}|"
    def set_author(self, author):
        self.__author = author
    def get_category(self):
        return f"Category: {self.category}"

                


class Movie(Product):
    category = 'Movie'
    def __init__ (self, item_id, title, price,quantity):
        self.price = price
        self.item_id = item_id
        self.quantity = quantity
        self.title = title
        self.__genre = None
        self.__year = None
    def get_info(self):
        return f"|{self.category}|\t{self.item_id}|\t{self.title}|\t{self.price}|\t{self.quantity}|\t{self.__year}|\t{self.__genre}|"     
    def get_category(self):
        return f"Category: {self.category}"
    def set_genre(self, genre):
        self.__genre = genre
    def set_year(self, year):
        self.__year = year
#     def lngth():
#         longest = 0
#         for item in self.__list:
#             if longest < len(item):
#                 longest = len(item)+1
#         for item in self.__list:
#             return item.rjust(longest)
    
    
class Inventory():
    def __init__ (self):
        self.__list = []
    def add_item (self, item):
        self.__list.append(item)
    def show_inventory(self):
        msg = ""
        index = 1
        for item in self.__list:
            msg += str(index) + ". "  + item.get_info() + "\n"
            index += 1 
        return msg




        
# book1 = Book(2190, "The Sun Also Rises", 12,100)

# movie1 = Movie(2102, "Alien", 23, 234)

# inventory_list = Inventory()

# inventory_list.add_item(book1)
# inventory_list.add_item(movie1)

# print(inventory_list.show_inventory())



```

```python
import pandas as pd 

#i realized halfway through doing this with pandas that we were probably supposed to use SQL... 
#but I was past the point of no return by then 


#import csv file as a dataframe, give it titles to reflect the weird mix of column data
df = pd.read_csv("productlist.csv", header = None, names = ["Category",'Item_ID', 'Title', 'Movie Genre/Book Author','Movie Year/Book Price','Movie Price/Book Quantity', 'Movie Quantity'])

df.head()

#drop the product unique values that won't go into the inventory
df.drop(columns='Movie Genre/Book Author', inplace=True)


#move the movie prices into the year column so that they match up
df.loc[df.Category == "MOVIE", "Movie Year/Book Price"] = df['Movie Price/Book Quantity']

#move the movie quantity into the same column the book quantity is in
df.loc[df.Category == "MOVIE", "Movie Price/Book Quantity"] = df['Movie Quantity']

#drop the last column
df.drop(columns='Movie Quantity', inplace=True)

#rename the columns 
df.columns = ['Category','Item_ID','Title','Price','Quantity']


df.head()
```

```python
#create new empty inventory list
inventory_list = Inventory()

#create new data frame with just the movie values
df2 = df.where(df['Category'] == 'MOVIE')
#drop empty columns
df2.dropna(inplace=True)
#turn the dataframe into a list
movielist = df2.values.tolist()

#iterate over the movie list to append to the inventory list
for i in movielist:
    m =  Movie(i[1], i[2], i[3],i[4])
    inventory_list.add_item(m)

    
#now do the same with the books 
df3 = df.where(df['Category'] == 'BOOK')
df3.dropna(inplace=True)
booklist = df3.values.tolist()

for i in booklist:
    b =  Book(i[1], i[2], i[3],i[4])
    inventory_list.add_item(b)


#print(inventory_list.show_inventory())
```
